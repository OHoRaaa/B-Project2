```python
import pandas as pd
import numpy as np
```

```python
import warnings

warnings.filterwarnings(action='ignore')
```

```python
def m_to_k(x):
    if x[-1] == 'M':
        return float(x[:-1]) * 1000
    elif x[-1] == 'K':
        return float(x[:-1])
    elif x[-1] == 'B':
        return float(x[:-1]) * 1000000
    else:
        return np.nan
```

```python
def change_date(x):
    ret = []
    for da in x.split():
        ret.append(da[:-1])
    return '.'.join(ret)
```

```python
# LIT ETF

df_lit = pd.read_csv('./Data/리튬 etf LIT.csv')
df_lit
```

```python
df_lit.drop(['Open', 'High', 'Low', 'Close'], axis = 1, inplace=True)
df_lit
```

```python
df_lit.columns = 'date', 'LIT조정종가', 'LIT거래량'
df_lit
```

```python
df_lit['date'] = pd.to_datetime(df_lit['date'])
```

```python
df_lit.info()
```

```python
# SGL ETF

df_sgl = pd.read_csv('./Data/리튬전지 SGL etf.csv')
df_sgl
```

```python
df_sgl.drop(['오픈', '고가', '저가', '거래량'], axis = 1, inplace = True)
df_sgl
```

```python
df_sgl.columns = 'date', 'SGL종가', 'SGL변동률'
df_sgl
```

```python
df_sgl['date'] = pd.to_datetime(df_sgl['date'].apply(change_date))
```

```python
df_sgl
```

```python
df_sgl['SGL변동률'] = df_sgl['SGL변동률'].apply(lambda x :x[:-1]).astype('float')
```

```python
df_sgl.info()
```

```python
df_sgl
```

```python
# merge

df_ls = pd.merge(df_lit, df_sgl, on = 'date', how = 'outer')
df_ls
```
