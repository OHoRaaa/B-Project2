```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score, mean_absolute_error
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoost
```

```python
y = df_tesla['tesla_volatility']
x = df_tesla.drop(['date','tesla', 'elec_fee', 'tesla_volatility',\
                   ], axis=1)
```

```python
tr = int(len(df_tesla) / 100 * 80)
x_train, x_test = x.iloc[1007:1600], x.iloc[1600:]
y_train, y_test = y.iloc[1007:1600], y.iloc[1600:]
```

```python
scaled = StandardScaler()
scaled.fit(x_train)
x_train_scaled = scaled.transform(x_train)
x_test_scaled = scaled.transform(x_test)
```

```python
lr = LinearRegression()
lr.fit(x_train_scaled, y_train)
preds_lr = lr.predict(x_test_scaled)
print('train :', lr.score(x_train_scaled,y_train))
print('test :', lr.score(x_test_scaled,y_test))
```

```python
rid = Ridge(alpha=100)
rid.fit(x_train_scaled, y_train)
preds_rid = rid.predict(x_test_scaled)
print('train :', rid.score(x_train_scaled,y_train))
print('test :', rid.score(x_test_scaled,y_test))
```

```python
las = Lasso(alpha=5)
las.fit(x_train_scaled, y_train)
preds_las = las.predict(x_test_scaled)
print('train :', las.score(x_train_scaled,y_train))
print('test :', las.score(x_test_scaled,y_test))
```

```python
xb = XGBRegressor()
xb.fit(x_train_scaled, y_train)
preds_xb = xb.predict(x_test_scaled)
print('train :', xb.score(x_train_scaled,y_train))
print('test :', xb.score(x_test_scaled,y_test))
```

```python
lgbm = LGBMRegressor()
lgbm.fit(x_train_scaled, y_train)
preds_lgbm = lgbm.predict(x_test_scaled)
print('train :', lgbm.score(x_train_scaled,y_train))
print('test :', lgbm.score(x_test_scaled,y_test))
```

```python
cat = CatBoost()
cat.fit(x_train_scaled, y_train)
preds_cat = cat.predict(x_test_scaled)
r2_score(y_test,preds_cat)
# print('train :', cat.score(x_train_scaled,y_train))
# print('test :', cat.score(x_test_scaled,y_test))
```

```python
y_log = np.log1p(y)
# log1p(n) = log(n+1) : 값이 0일때 error를 해결하기 위해
y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)
```

```python
y_log.hist(bins=50, figsize=(10,5))
```

```python
lr = LinearRegression()
rid = Ridge()
las = Lasso(alpha=0.0001)

lr.fit(x_train_scaled, y_train_log)
rid.fit(x_train_scaled, y_train_log)
las.fit(x_train_scaled, y_train_log)

preds_lr = lr.predict(x_test_scaled)
preds_rid = rid.predict(x_test_scaled)
preds_las = las.predict(x_test_scaled)

# exponential을 통해 log원복
preds_lr_exp = np.expm1(preds_lr)
preds_rid_exp = np.expm1(preds_rid)
preds_las_exp = np.expm1(preds_las)

# log를 취해 쏠린 label값을 조정해 준 것이 오차가 더 작다.
print('r2_score_lr :', r2_score(y_test, preds_lr_exp))
print('r2_score_rid :', r2_score(y_test, preds_rid_exp))
print('r2_score_las :', r2_score(y_test, preds_las_exp))
```

```python
df_ = df_f.iloc[1027:]
```

```python
y = df_['tesla']
x = df_.drop(['tesla_volatility', 'date', 'tesla', 'elec_fee'], axis = 1)
```

```python
x_train, x_test = x.iloc[:560], x.iloc[560:]
y_train, y_test = y.iloc[:560], y.iloc[560:]
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
preds_lr = lr.predict(x_test)
print('train :', lr.score(x_train,y_train))
print('test :', lr.score(x_test,y_test))
```

```python
# 스케일링

scaled = StandardScaler()
scaled.fit(x_train)
x_train_scaled = scaled.transform(x_train)
x_test_scaled = scaled.transform(x_test)
```

```python
# LinearRegression 

lr = LinearRegression()
lr.fit(x_train_scaled, y_train)
preds_lr = lr.predict(x_test_scaled)
print('train :', lr.score(x_train_scaled,y_train))
print('test :', lr.score(x_test_scaled,y_test))
```

```python
preds_lr
```

```python
# Ridge

train_score = []
test_score = []
alpha_list = list(np.arange(0.001, 100, 0.001))

for alpha in alpha_list:
    rid = Ridge(alpha=alpha)
    rid.fit(x_train_scaled, y_train)
    train_score.append(rid.score(x_train_scaled, y_train))
    test_score.append(rid.score(x_test_scaled, y_test))
    preds_rid = rid.predict(x_test_scaled)
print(train_score)
print(test_score)
```

```python
preds_rid
```

```python
plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.show()
```

```python
rid = Ridge(alpha=0.01)
rid.fit(x_train_scaled, y_train)
print(rid.score(x_train_scaled, y_train))
print(rid.score(x_test_scaled, y_test))
```

```python
# Lasso

train_score = []
test_score = []
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]

for alpha in alpha_list:
    las = Lasso(alpha=alpha)
    las.fit(x_train_scaled, y_train)
    train_score.append(las.score(x_train_scaled, y_train))
    test_score.append(las.score(x_test_scaled, y_test))
    preds_las = las.predict(x_test_scaled)
print(train_score)
print(test_score)
```

```python
plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.show()
```

```python
las = Lasso(alpha=1)
las.fit(x_train_scaled, y_train)
preds = las.predict(x_test_scaled)
print(las.score(x_train_scaled, y_train))
print(las.score(x_test_scaled, y_test))
```

```python
max_score = 0
max_alpha = 0
max_ratio = 0
max_train_score = 0

for rat in np.arange(0.1,1,0.01):
    for alp in np.arange(1,15,0.1):
        
        ell = ElasticNet(l1_ratio=rat, alpha=alp)
        ell.fit(x_train_scaled, y_train)
        preds_ell = ell.predict(x_test_scaled)
        if max_score < ell.score(x_test_scaled, y_test):
            max_score = ell.score(x_test_scaled, y_test)
            max_alpha = alp
            max_ratio = rat
            max_train_score = ell.score(x_train_scaled, y_train)
print('alpha :', max_alpha)
print('ratio :', max_ratio)
print('train_score :' , max_train_score)
print('test_score :', max_score)
```
